```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
import os

os.chdir("..")  # Moverse desde /notebooks al directorio raíz del proyecto
print("Directorio actual:", os.getcwd())

df = pd.read_pickle("data/intermediate/df.pkl")

categorical_cols = ['Gender','Caste','coaching','time','Class_ten_education','twelve_education','medium','Father_occupation','Mother_occupation']
X = df[[c+'_freq' for c in categorical_cols]+['Academic_Score']]
y = LabelEncoder().fit_transform(df['Performance_grouped'])

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

import pickle
with open("data/processed/Xy_resampled.pkl", "wb") as f:
    pickle.dump((X_res, y_res), f)
```
