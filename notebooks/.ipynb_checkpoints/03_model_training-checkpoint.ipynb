```python
# -- Librerias -- #
import pickle
from pathlib import Path
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from datetime import datetime
import time

from sklearn.model_selection import StratifiedKFold, GridSearchCV, train_test_split
from sklearn.metrics import make_scorer, f1_score, classification_report, confusion_matrix

from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from sklearn.ensemble import ExtraTreesClassifier


# -- Configuracion de rutas -- #
root_dir = Path.cwd().parents[0]    # Ruta del proyecto
processed_data = root_dir / "data" / "processed"   # Ruta de la data que usaremos
model_dir = root_dir / "models"    # Ruta donde meteremos los modelos
model_summary_dir = root_dir / "models_summary"    # Ruta donde meteremos los resultados de los modelos

# -- Carga de datos train test -- #
with open(processed_data / "Xy_train_resampled.pkl", "rb") as f:
    X_train, y_train = pickle.load(f)

with open(processed_data / "Xy_test.pkl", "rb") as f:
    X_test, y_test = pickle.load(f)


# -- Definir modelos e hiperparametros -- #

models = {
    'RandomForest': {
        'model': RandomForestClassifier(random_state=42, class_weight='balanced'),
        'params': {
            'n_estimators': [300, 350],
            'max_depth': [5, 10],
            'min_samples_leaf': [3, 5],
            'min_samples_split': [10, 12, 15],
            'max_features': ['sqrt', 'log2'],
            'bootstrap': [True, False]
        }
    },
    'XGBoost': {
        'model': XGBClassifier(objective='multi:softprob', eval_metric='mlogloss', random_state=42, tree_method='hist', use_label_encoder=False,
            n_jobs=-1),
        'params': {
            'n_estimators': [100, 150],       # no tan alto, evita sobreajuste
            'max_depth': [3, 4],              # controla complejidad
            'learning_rate': [0.05, 0.1],     # trade-off entre velocidad y generalización
            'subsample': [0.7, 0.9],          # para evitar sobreajuste
            'colsample_bytree': [0.7, 0.9],   # más diversidad en los árboles
            'gamma': [0, 1],                  # regularización en splits
            'reg_alpha': [0, 0.1],            # L1 regularization
            'reg_lambda': [1, 1.5],           # L2 regularization
            'min_child_weight': [1, 3, 5]     # evita árboles muy profundos en clases pequeñas
        }
    },
    'CatBoost': {
        'model': CatBoostClassifier(iterations=300, verbose=0, random_seed=42, class_weights=[1, 2, 2]),
        'params': {
            'depth': [4, 6],
            'learning_rate': [0.05, 0.07],
            'l2_leaf_reg': [1, 3, 5],
            'border_count': [64]
        }
    },
    'ExtraTrees': {
        "model": ExtraTreesClassifier(random_state=42, n_jobs=-1, class_weight='balanced'),
        "params": {
           "n_estimators": [200, 400],
            "max_depth": [5, 10, None],
            "min_samples_split": [2, 5],
            "min_samples_leaf": [1, 3],
            "max_features": ["sqrt", "log2"]
        }
    }
}

# -- Funcion de entrenamiento con GridSearchCV -- #
def train_model(model, params, X, y, cv, scorer):
    gs = GridSearchCV(model, params, cv=cv, scoring=scorer, n_jobs=-1)
    gs.fit(X, y)
    return gs.best_estimator_, gs.best_score_, gs.best_params_

# -- Cross-Validation y entrenamiento -- #
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
f1_scorer = make_scorer(f1_score, average='macro')
best_models = {}
results_summary = []

for name, m in models.items():
    print(f"\n--- Entrenando {name} ---")
    # Inicio de Entrenamiento - Tiempo 
    start_time = time.time()
    best_model, best_score, best_params = train_model(m['model'], m['params'], X_train, y_train, cv, f1_scorer)
    print(f"Mejor F1 (CV): {best_score:.3f}")
    print("Mejores parámetros:", best_params)

    # Fin de Entrenamiento - Tiempo
    end_time = time.time()
    elapsed = end_time - start_time
    print(f"Tiempo de entrenamiento: {elapsed:.2f} segundos")
    
    # Evaluación final sobre test set
    y_pred = best_model.predict(X_test)
    print("\nClassification Report (Test set):")
    print(classification_report(y_test, y_pred))

    # Definir thresholds personalizados por clase (0: mayoritaria, 1-3: minoritarias)
    # Puedes ajustar estos valores según experimentes con recall
    thresholds = {
        0: 0.5,   # clase mayoritaria, se mantiene estándar
        1: 0.32,  # clase minoritaria, se reduce threshold
        2: 0.25,  # clase minoritaria, se reduce threshold
        # 3: 0.30   # clase minoritaria, se reduce threshold
    }

# --- Ajuste de predicciones basado en thresholds ---
    if hasattr(best_model, "predict_proba"):
        probs = best_model.predict_proba(X_test)
        y_pred_adj = []
        for p in probs:
            # Buscar todas las clases que superan su threshold
            candidates = [i for i, prob in enumerate(p) if prob >= thresholds[i]]
            if candidates:
                # Escoger la de mayor probabilidad entre las candidatas
                y_pred_adj.append(candidates[np.argmax([p[i] for i in candidates])])
            else:
                # Ninguna supera threshold, escoger la de mayor probabilidad
                y_pred_adj.append(np.argmax(p))
        y_pred_adj = np.array(y_pred_adj)
    else:
            # Para modelos que no soportan predict_proba
            y_pred_adj = y_pred

    # --- Evaluación final con thresholds ajustados ---
    print("\nClassification Report (Test set) con thresholds ajustados:")
    print(classification_report(y_test, y_pred_adj))

    # Matriz de confusión normalizada con thresholds ajustados
    cm = confusion_matrix(y_test, y_pred_adj)
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.figure(figsize=(6,5))
    sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap="Blues")
    plt.title(f'Matriz de confusión normalizada (Thresholds) - {name}')
    plt.ylabel("True")
    plt.xlabel("Pred")
    plt.show()
        
    # Guardado del modelo
    timestamp = datetime.now().strftime("%Y%m%d_%H%M")
    model_path = model_dir / f"{name}_best_model_{timestamp}.pkl"
    joblib.dump(best_model, model_path)
    
    print(f"Modelo guardado en: {model_path}")
    
    best_models[name] = best_model
    results_summary.append({
        "Model": name,
        "F1_CV": best_score,
        "F1_Test": f1_score(y_test, y_pred, average='macro'),
        "F1_Test_Threshold": f1_score(y_test, y_pred_adj, average='macro'),
        "Train_Time_sec": round(elapsed, 2)
    })

# -- Resumen de Resultados -- #
results_df = pd.DataFrame(results_summary).sort_values(by="F1_Test_Threshold", ascending=False)
results_df = results_df.round(3)
results_df.to_csv(model_summary_dir / f"results_summary_{timestamp}.csv", index=False)

print("\nResumen comparativo de modelos:")
display(results_df)
```
